## Carga de datos y librerias


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import joblib

try:
    df = pd.read_csv('Loan_Default.csv')
except FileNotFoundError:
    print("'Loan_Default.csv' esté subido a tu sesión de Colab.")
    df = pd.DataFrame()

if not df.empty:
  print("### Datos Iniciales (HEAD)")
  print(df.head())
  print("\n### Información del DataFrame (INFO)")
  print(df.info())
```

## Comprension de Datos (EDA)

```python
if not df.empty:
    TARGET_COL = 'Status'

    print("\n### Análisis de la Variable Objetivo: Desbalance de Clases")
    print(df[TARGET_COL].value_counts(normalize=True))
    sns.countplot(x=TARGET_COL, data=df)
    plt.title('Distribución de Clases (0: No Default, 1: Default)')
    plt.show()

    print("\n### Conteo de Valores Nulos por Columna")
    print(df.isnull().sum()[df.isnull().sum() > 0])

    numeric_cols = df.select_dtypes(include=np.number).columns
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Matriz de Correlación de Variables Numéricas')
    plt.show()
```

## Preparacion de Datos

```python
if not df.empty:
    # 3.1. Limpieza de Datos

    # (Paso 1) Eliminar columnas irrelevantes
    columns_to_drop = ['ID', 'year']
    df = df.drop(columns=columns_to_drop, errors='ignore')

    # (Paso 2) Eliminar filas duplicadas (si existen)
    print(f"Filas iniciales: {df.shape[0]}")
    df.drop_duplicates(inplace=True)
    print(f"Filas después de eliminar duplicados: {df.shape[0]}")

    # (Paso 3) Ingeniería de Variables y Limpieza Específica (Opcional)

    # --- 3.2. División en Conjuntos de Entrenamiento y Prueba ---
    

    X = df.drop(TARGET_COL, axis=1)
    y = df[TARGET_COL]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\n### Dimensiones de los conjuntos (Train/Test)")
    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"Proporción de 'Default' en Entrenamiento: {y_train.mean():.4f}")

    # --- 3.3. Creación del Pipeline de Preprocesamiento (Transformación de Variables) ---

    print("\n### Creación del Pipeline de Preprocesamiento")

    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    NUMERIC_FEATURES = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

    CATEGORICAL_FEATURES = X_train.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

    print(f"Columnas numéricas ({len(NUMERIC_FEATURES)}): {NUMERIC_FEATURES}")
    print(f"Columnas categóricas ({len(CATEGORICAL_FEATURES)}): {CATEGORICAL_FEATURES}")

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_pipeline, NUMERIC_FEATURES),
            ('cat', categorical_pipeline, CATEGORICAL_FEATURES)
        ],
        remainder='drop'
    )

    print("Preprocesador (ColumnTransformer) listo. Se aplica a las siguientes features:")
    print(f"  Numéricas ({len(NUMERIC_FEATURES)}): {NUMERIC_FEATURES}")
    print(f"  Categóricas ({len(CATEGORICAL_FEATURES)}): {CATEGORICAL_FEATURES}")
```

## Modelado y Evaluacion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
import joblib

# --- Asumiendo que df, X_train, y_train, X_test, y_test, y preprocessor ya están definidos ---
# Por ejemplo, si X_train.shape = (104069, 31)

if 'df' in locals() and not df.empty:
    
    # --- Modificación para Acelerar la Búsqueda de Hiperparámetros (Fase 4) ---
    
    # 1. Crear una Muestra Reducida del 10% de X_train y y_train
    # Esto reduce la complejidad de SVM de manera drástica.
    
    # Nota: Si el error 'df' is not defined persiste, este bloque no se ejecutará.
    # Se añade un try/except para definir X_train/y_train para este ejemplo
    try:
        # Se requiere importar train_test_split nuevamente si no estaba en la cabecera
        from sklearn.model_selection import train_test_split
        
        # Usar solo el 10% del conjunto de entrenamiento (~10,400 filas)
        X_sample, _, y_sample, _ = train_test_split(
            X_train, y_train, 
            test_size=0.9, # Descartar el 90%
            random_state=42, 
            stratify=y_train 
        )
        
        print(f"Usando muestra reducida para el ajuste de hiperparámetros: {X_sample.shape[0]} filas.")
        print(f"Esto reduce el tamaño de {X_train.shape[0]} a {X_sample.shape[0]} filas.")

    except NameError:
        print("Variables X_train o y_train no definidas. No se puede crear la muestra. Usando el conjunto completo.")
        X_sample, y_sample = X_train, y_train # Fallback (Lento)
    except Exception as e:
        print(f"Error al muestrear: {e}. Usando el conjunto completo.")
        X_sample, y_sample = X_train, y_train # Fallback (Lento)

    # 4.1. Definición de Modelos (KNN y SVM)

    # 4.2. Pipeline completo para KNN
    knn_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', KNeighborsClassifier())
    ])

    # 4.3. Pipeline completo para SVM
    svm_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', SVC(probability=True, random_state=42)) 
    ])

    # 4.4. Ajuste de Hiperparámetros (GridSearch)

    # Parámetros para KNN
    param_grid_knn = {
        'classifier__n_neighbors': [5, 10, 15],
        'classifier__weights': ['uniform', 'distance']
    }

    # Parámetros para SVM (Se usa un rango pequeño por la complejidad computacional)
    param_grid_svm = {
        'classifier__C': [0.1, 1],
        'classifier__kernel': ['linear', 'rbf']
    }


    print("\n### 4. Modelado: Entrenamiento y Ajuste de Hiperparámetros (KNN)")
    grid_search_knn = GridSearchCV(knn_pipeline, param_grid_knn, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=1)
    # AJUSTE CON LA MUESTRA REDUCIDA
    grid_search_knn.fit(X_sample, y_sample) 
    best_knn = grid_search_knn.best_estimator_

    print("\n### 4. Modelado: Entrenamiento y Ajuste de Hiperparámetros (SVM)")
    grid_search_svm = GridSearchCV(svm_pipeline, param_grid_svm, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=1)
    # AJUSTE CON LA MUESTRA REDUCIDA
    grid_search_svm.fit(X_sample, y_sample)
    best_svm = grid_search_svm.best_estimator_


    # -----------------------------------------------------------
    # 5. Evaluación (Usa el conjunto de PRUEBA COMPLETO: X_test, y_test)
    # -----------------------------------------------------------

    models = {
        'KNN': best_knn,
        'SVM': best_svm
    }

    results = {}

    print("\n### 5. Evaluación: Resultados en el Conjunto de Prueba")
    for name, model in models.items():
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

        report = classification_report(y_test, y_pred, output_dict=True)
        auc_score = roc_auc_score(y_test, y_proba)

        results[name] = {
            'Accuracy': accuracy_score(y_test, y_pred),
            'Recall (Default)': report['1']['recall'],
            'F1-Score (Default)': report['1']['f1-score'],
            'AUC-ROC': auc_score,
            'Report': report
        }

        print(f"\n--- Resultados para {name} ---")
        print(f"Mejores Parámetros: {model.named_steps['classifier'].get_params()}")
        print(classification_report(y_test, y_pred))
        print(f"AUC-ROC: {auc_score:.4f}")
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
        plt.title(f'Matriz de Confusión {name}')
        plt.show()

    # 5.2. Conclusión del Modelo (Comparación Final)
    results_df = pd.DataFrame(results).T
    print("\n### Comparación Final de Modelos")
    print(results_df[['Accuracy', 'Recall (Default)', 'F1-Score (Default)', 'AUC-ROC']])

    best_model_name = results_df['Recall (Default)'].idxmax()
    best_model = models[best_model_name]
    print(f"\n✅ El **Mejor Modelo** (basado en Recall para la clase Default) es: **{best_model_name}**.")


    # 5.3. Guardar el Mejor Modelo para el Despliegue (Fase 6)
    joblib.dump(best_model, 'best_credit_risk_model.pkl')
    print("\nModelo final guardado como 'best_credit_risk_model.pkl' para el despliegue.")
```
